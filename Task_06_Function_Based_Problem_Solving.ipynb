```python
# =====================================================
# Problem 1: Smart Parking Lot Management System
# =====================================================

def parking_status(capacity, logs):
    parked = 0
    peak = 0

    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1

        peak = max(peak, parked)

    status = "Full" if parked >= capacity else "Available"

    print("Currently Parked Vehicles:", parked)
    print("Peak Usage:", peak)
    print("Parking Status:", status)


# Sample Input
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

parking_status(capacity, vehicle_logs)
```

```python
# =====================================================
# Problem 2: Online Food Delivery Time Estimator
# =====================================================

def estimate_delivery_time(distance, traffic, weather):
    base_time = distance * 5  # 5 min per km

    if traffic.lower() == "high":
        base_time += 15
    elif traffic.lower() == "medium":
        base_time += 8

    if weather.lower() == "rainy":
        base_time += 10
    elif weather.lower() == "storm":
        base_time += 20

    print("Estimated Delivery Time:", base_time, "minutes")


# Sample Input
estimate_delivery_time(8, "High", "Rainy")
```

```python
# =====================================================
# Problem 3: Movie Theatre Seat Occupancy Analyzer
# =====================================================

def analyze_occupancy(total_seats, booked_list):
    booked = len(booked_list)
    occupancy = (booked / total_seats) * 100

    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Available"

    print(f"Occupancy: {occupancy:.0f}%")
    print("Show Status:", status)


# Sample Input
booked_seats = [1] * 150
analyze_occupancy(200, booked_seats)
```

```python
# =====================================================
# Problem 4: Cloud Server Load Classification
# =====================================================

def classify_server_load(cpu_readings):
    avg_cpu = sum(cpu_readings) / len(cpu_readings)

    if avg_cpu < 50:
        status = "Normal"
    elif avg_cpu <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print(f"Average CPU Load: {avg_cpu:.0f}%")
    print("Server Status:", status)


# Sample Input
classify_server_load([45, 60, 70, 85, 90])
```

```python
# =====================================================
# Problem 5: Smart Classroom Resource Usage Monitor
# =====================================================

def resource_monitor(usage_dict):
    threshold = 8  # hours
    overused = []

    for resource, hours in usage_dict.items():
        if hours > threshold:
            overused.append(resource)

    print("Overused Resources:", ", ".join(overused) if overused else "None")
    print("Energy Alert:", "Yes" if overused else "No")


# Sample Input
usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

resource_monitor(usage)
```

```python
# =====================================================
# Problem 6: Online Event Registration Controller
# =====================================================

def registration_control(capacity, registrations):
    if registrations <= capacity:
        confirmed = registrations
        waitlist = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlist = registrations - capacity
        status = "Closed"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)


# Sample Input
registration_control(100, 105)
```
