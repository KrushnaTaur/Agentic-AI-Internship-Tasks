```python
# =====================================================
# Problem 1: Smart Light Controller
# =====================================================

class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def display_status(self):
        print(f"{self.name} is {self.status}")


# Sample Input
light = SmartLight("Bedroom Light")
light.turn_on()
light.display_status()
```

```python
# =====================================================
# Problem 2: Employee ID Card System
# =====================================================

class EmployeeID:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# Sample Input
emp = EmployeeID("Rahul", "EMP102", "IT")
emp.display_card()
```

```python
# =====================================================
# Problem 3: Mobile Contact Record
# =====================================================

class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# Sample Input
contact = Contact("Anita", 9876543210)
contact.display_contact()
```

```python
# =====================================================
# Problem 4: Product Price Tag Generator
# =====================================================

class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display_tag(self):
        print("Product:", self.name)
        print(f"Price: ₹{self.price}")


# Sample Input
product = Product("Headphones", 2499)
product.display_tag()
```

```python
# =====================================================
# Problem 5: Movie Rating Display System
# =====================================================

class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_movie(self):
        print("Movie:", self.name)
        print(f"Rating: {self.rating} / 5")


# Sample Input
movie = Movie("Inception", 4.8)
movie.display_movie()
```

```python
# =====================================================
# Problem 6: Delivery Address Manager
# =====================================================

class Delivery:
    def __init__(self, customer, address):
        self.customer = customer
        self.address = address

    def display_details(self):
        print("Delivery Details")
        print("Customer:", self.customer)
        print("Address:", self.address)


# Sample Input
delivery = Delivery("Suman", "Hyderabad")
delivery.display_details()
```
